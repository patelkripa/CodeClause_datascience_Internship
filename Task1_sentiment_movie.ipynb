```python
# Step 8: Evaluate Model
y_pred = model.predict(X_test)
print(" Accuracy:", accuracy_score(y_test, y_pred))
print(" Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import streamlit as st
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression  # or from sklearn.svm import LinearSVC
import joblib

# ----------------------------
# 1. Text Preprocessing
# ----------------------------
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"http\S+|www.\S+", "", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# ----------------------------
# 2. Load model and vectorizer
# ----------------------------
@st.cache_resource
def load_model():
    model = joblib.load("sentiment_model.pkl")
    vectorizer = joblib.load("tfidf_vectorizer.pkl")
    return model, vectorizer

model, vectorizer = load_model()

# ----------------------------
# 3. Streamlit App Interface
# ----------------------------
st.title(" IMDB Movie Review Sentiment Classifier")
st.write("Enter a movie review below and the model will classify it as **Positive** or **Negative**.")

review = st.text_area("Enter your movie review here", height=200)

if st.button("Predict Sentiment"):
    if review.strip() == "":
        st.warning("Please enter a review.")
    else:
        clean_review = clean_text(review)
        X_input = vectorizer.transform([clean_review])
        prediction = model.predict(X_input)[0]

        if prediction == 1:
            st.success(" Positive Review!")
        else:
            st.error(" Negative Review.")

```
